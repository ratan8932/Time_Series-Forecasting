```python
# Importing Libraries
import pandas as pd
import numpy as np
from pandas import read_csv
from matplotlib import pyplot as plt
from pandas.plotting import lag_plot
from statsmodels.graphics.tsaplots import plot_acf
from pandas import DataFrame
from pandas import Grouper
```

```python
series = read_csv('daily-minimum-temperatures.csv', header = 0, index_col=0, parse_dates=True)
series
```

### Line plot

```python
# Creat a Line plot
series.plot(figsize=(16,9))
plt.show()
```

### Histogram and Density Plots

```python
series.hist(figsize=(12,8))
plt.show()
```

```python
# create a density plot
from pandas import read_csv
from matplotlib import pyplot
series.plot(kind='kde', figsize=(12,8))
plt.show()
```

#### Box and Whisker Plots by Interval

```python
# create a boxplot of yearly data

groups = series.groupby(Grouper(freq='A'))
years = DataFrame()
for name, group in groups:
    years[name.year] = (group.values.flatten())
years
```

```python
years.boxplot(figsize=(12,8))
plt.show()
```

### Lag plot

```python
# create a scatter plot
for i in [1, 30, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 365]:
    #plt.figure(figsize=(12,8))
    lag_plot(series, lag=i)
    plt.show()
```

### Autocorrealtion plot

```python
# create an autocorrelation plot
n, m = 12, 8

with plt.rc_context():
    plt.rc("figure", figsize=(n,m))

    plt.figure(figsize=(12,8))
    plot_acf(series,lags=365)
    plt.show()
```
